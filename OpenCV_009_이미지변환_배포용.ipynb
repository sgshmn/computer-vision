# 이미지 변환

## 변환의 종류
 
  - 강체 변환(Ridid-Body transformation) : 이동, 회전
  - 유사 변환(Similarity Transformations) : 강체 변환 후 크기 변환
  - 선형 변환(Linear Transformation)
  - 원근변환(Projective Transformation)
  
  - Affine Transformations : (강체변환 + 유사변환) + 선형 변환 - 이동, 회전, 크기, 기울림, 반전 변환

## 이미지 확대

- <font color=red>cv2.resize(src, (w, h), interpolation)</font> 
  - src를 w, h 만큼 크기로 변환
  
  
- interpolation : 보간법 (이미지 변환시 깨지는 부분을 보간)
  - cv2.INTER_NEAREST : 최근방 이웃 보간법 - 빠르지만 화질이 떨어짐, 잘 쓰이지 않음
  - cv2.INTER_LINEAR : 양선형 보간법 (2x2 이웃 픽셀 참조) - 효율이 가장 좋음, 속도도 빠르고 화질도 우수
  - cv2.INTER_CUBIC : 3차 회선 보간법 (4x4 이웃 픽셀 참조) - NTER_LINEAR보다 느리지만 화질이 더 우수
  - cv2.INTER_LANCZOS4 : Lanczos 보간법 (8x8 이웃 픽셀 참조) - 속도가 느리지만 화질이 더 우수
  - cv2.INTER_AREA : 영역적인 정보를 추출해서 결과 영상을 셋팅 - 영상 축소시 효과적인 방법 

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread('images/flower1.png', cv2.IMREAD_COLOR)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.subplot(2, 3, 1)
plt.imshow(img1)
plt.axis('off')
plt.title('ORIGINAL')


h, w, _ = img1.shape



img2 = cv2.resize(img1, (w * 3, h * 3), interpolation=cv2.INTER_NEAREST)
plt.subplot(2, 3, 2)
plt.imshow(img2)
plt.axis('off')
plt.title('NEAREAST')



img3 = cv2.resize(img1, (w * 3, h * 3), interpolation=cv2.INTER_LINEAR)
plt.subplot(2, 3, 3)
plt.imshow(img3)
plt.axis('off')
plt.title('LINEAR')



img4 = cv2.resize(img1, (w * 3, h * 3), interpolation=cv2.INTER_CUBIC)
plt.subplot(2, 3, 4)
plt.imshow(img4)
plt.axis('off')
plt.title('CUBIC')

img5 = cv2.resize(img1, (w * 3, h * 3), interpolation=cv2.INTER_LANCZOS4)
plt.subplot(2, 3, 5)
plt.imshow(img5)
plt.axis('off')
plt.title('LANCZOS4')

img6 = cv2.resize(img1, (w * 3, h * 3), interpolation=cv2.INTER_AREA)
plt.subplot(2, 3, 6)
plt.imshow(img6)
plt.axis('off')
plt.title('AREA')
```

## 이미지 축소

- 이미지 축소 시 한 픽셀로 구성된 성분은 사라지게 되는 경우가 발생
  - 입력 영상을 부드럽게 필터링한 후에 축소를 하거나 여러번 축소를 반복 (cv2.INTER_AREA)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img1 = cv2.imread('images/shape.png', cv2.IMREAD_COLOR)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

plt.subplot(2, 3, 1)
plt.imshow(img1)
plt.axis('off')
plt.title('ORIGINAL')


h, w, _ = img1.shape

img2 = cv2.resize(img1, (int(w * 0.2), int(h * 0.2)), interpolation=cv2.INTER_NEAREST)
plt.subplot(2, 3, 2)
plt.imshow(img2)
plt.axis('off')
plt.title('NEAREAST')



img3 = cv2.resize(img1, (int(w * 0.2), int(h * 0.2)), interpolation=cv2.INTER_LINEAR)
plt.subplot(2, 3, 3)
plt.imshow(img3)
plt.axis('off')
plt.title('LINEAR')



img4 = cv2.resize(img1, (int(w * 0.2), int(h * 0.2)), interpolation=cv2.INTER_CUBIC)
plt.subplot(2, 3, 4)
plt.imshow(img4)
plt.axis('off')
plt.title('CUBIC')

img5 = cv2.resize(img1, (int(w * 0.2), int(h * 0.2)), interpolation=cv2.INTER_LANCZOS4)
plt.subplot(2, 3, 5)
plt.imshow(img5)
plt.axis('off')
plt.title('LANCZOS4')

img6 = cv2.resize(img1, (int(w * 0.2), int(h * 0.2)), interpolation=cv2.INTER_AREA)
plt.subplot(2, 3, 6)
plt.imshow(img6)
plt.axis('off')
plt.title('AREA')
```

## 이미지 이동

- <font color=red>cv2.warpAffine(img, M, (w, h), dst, flags, borderValue)</font>
  - M : 행렬에 따른 affine 변환을 수행
  - (w, h) : 출력 이미지의 크기 
  - dst : 출력 영상에 대한 옵션
  - flags : 보간법 (cv2.INTER_LINEAR 등)
  - borderValue : 이동 후 빈공간 채우는 방식 (0 (검정색), cv2.BORDER_REFLECT(반사)

```python
img = cv2.imread('images/lenna.png', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('original')
plt.axis('off')

h, w, _ = img.shape

# 변환 행렬 (왼쪽으로 20, 위쪽으로 20 픽셀 이동)
M = np.float32([[1, 0, -100], # 왼쪽
               [0, 1, -20]]) # 위쪽

result = cv2.warpAffine(img1, M, (w, h), None, cv2.INTER_LINEAR, cv2.BORDER_REFLECT)


plt.subplot(1, 2, 2)
plt.imshow(result)
plt.title('move')
plt.axis('off')
```

## 이미지 회전

- <font color=red>cv2.getRotationMatrix2D(회전중심점, 각도, 확대비율)</font>

```python
# 회전 설정
M = cv2.getRotationMatrix2D((w/2, h/2), 20, 1)

result = cv2.warpAffine(img1, M, (w, h), None, cv2.INTER_LINEAR, cv2.BORDER_REFLECT)

plt.imshow(result)
plt.title('rotate')
plt.axis('off')
```

## 이미지 대칭

- <font color=red>cv2.flip(img, flipCode)</font>
  - flipCode : 대칭 방법
    - flipCode < 0 : 대각 대칭 
    - flipCode = 0 : 상하 대칭
    - flipCode > 0 : 좌우 대칭

```python
result = cv2.flip(img, -1)

plt.imshow(result)
plt.title('flip')
plt.axis('off')
```

## 원근 변환

- Affine 변환과는 달리 4개 점을 이용하여 이미지를 변환


- <font color=red>M = cv2.getPerspectiveTransform(pts1, pts2)</font>
  - pts1 좌표를 pts2 좌표로 원근 변환 설정


- <font color=red>result = cv2.warpPerspective(img, M, (w, h))</font>
  - 이미지, 변환값, 이미지 크기로 원근 변환 실행
  - (w, h) : 출력 이미지 크기

```python
# 이전 좌표
pts1 = np.float32([[0, 0], [w, 0], [0, h], [w, h]])

# 이후 좌표
pts2 = np.float32([[100, 100], [w-100, 100], [0, h], [w, h]])

M = cv2.getPerspectiveTransform(pts1, pts2)

result = cv2.warpPerspective(img, M, (w, h))

plt.imshow(result)
plt.title('projection')
plt.axis('off')
```

# 이미지 증식하기

- Image augmentation : 훈련 데이터가 적은 경우 훈련 데이터를 늘리는 방식

```python
import cv2
import matplotlib.pyplot as plt
import os
import random
import numpy as np
input_dir = "./images/augmentation/"
output_dir = "./images/output/"
if not os.path.exists(input_dir) :
    os.mkdir(input_dir)
if not os.path.exists(output_dir) :
    os.mkdir(output_dir)
# 이미지 불러오기
file_list = os.listdir(input_dir)
# 증식 횟수
epoch = 2
for file in file_list :
    img = cv2.imread(input_dir + file)
    file_name = file.split(".")[0]
    h, w, ch = img.shape
    # 이미지 증식
    for i in range(epoch) :
        # 이미지 확대 / 축소 값 선택
        param = random.uniform(0, 3)
        # 확대
        if param >= 1.0 :
            img2 = cv2.resize(img, (int(w * param), int(h * param)),
                     interpolation = cv2.INTER_CUBIC)
        # 축소
        else :
            img2 = cv2.resize(img, (int(w * param), int(h * param)),
                     interpolation = cv2.INTER_AREA)
        # 이미지 좌우 이동 선택
        param1 = random.randint(0, 1)
        # 이동 값
        param2 = random.randint(-10, 10)
        h, w, ch = img2.shape
        if param1 == 0 :   # 왼쪽 이동
            M = np.float32([[1, 0, param2],
                            [0, 1, 0]])
            img3 = cv2.warpAffine(img2, M, (w, h), None, cv2.INTER_LINEAR, cv2.BORDER_REFLECT)
        else :  # 오른쪽 이동
            M = np.float32([[1, 0, param2],
                            [0, 1, 0]])
            img3 = cv2.warpAffine(img2, M, (w, h), None, cv2.INTER_LINEAR,
                                  cv2.BORDER_REFLECT)
        # 회전값
        param = random.randint(-20, 20)
        h, w, ch = img3.shape
        if param == 0 :   # 왼쪽 회전
            M = cv2.getRotationMatrix2D((w / 2, h / 2), param, 1)
            img4 = cv2.warpAffine(img3, M, (w, h), None, cv2.INTER_LINEAR,
                                  cv2.BORDER_REFLECT)
        else :  # 오른쪽 회전
            M = cv2.getRotationMatrix2D((w / 2, h / 2), param, 1)
            img4 = cv2.warpAffine(img3, M, (w, h), None, cv2.INTER_LINEAR,
                                  cv2.BORDER_REFLECT)
        # 증식 이미지 저장
        cv2.imwrite(output_dir + "{}_{}.png".format(file_name, i), img4,
                    params=[cv2.IMWRITE_PNG_COMPRESSION, 0])
```

```python
plt.figure(figsize=(10, 5 * epoch))
co = 0
file_list = os.listdir(output_dir)
for file in file_list :
    co += 1
    img = cv2.imread(output_dir + file)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(epoch + 1, 3, co)
    plt.imshow(img)
```
